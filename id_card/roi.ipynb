```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret , img = cap.read()
    img = cv2.flip(img, 1)
    height= img.shape[0]
    width = img.shape[1]
    
    b_y1 = int(width*0.2)
    b_y2 = int(width*0.8)
    b_x1 = int(height*0.2)
    b_x2 = int(height*0.8)
    roi_box = [(b_y1,b_x1),(b_y2,b_x2)]
    roi_height = int(height*0.4)
    roi_width = int(width*0.4)


    # i_y1 = int(width*0.15)
    # i_y2 = int(width*0.85)
    # i_x1 = int(height*0.15)
    # i_x2 = int(height*0.85)
    # roi_img = img[i_x1:i_x2,i_y1:i_y2]


    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _,thresh1 = cv2.threshold(gray, 160,255, cv2.THRESH_BINARY)

    #contours, hierarchy = cv2.findContours(thresh1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    cnt, _ = cv2.findContours(thresh1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for cnt in cnt:
        area = cv2.contourArea(cnt)
        if area > 7000:
            rect = cv2.minAreaRect(cnt)
            box = cv2.boxPoints(rect)
            box = np.int0(box)
            cv2.drawContours(img, [box], 0, (0, 255, 0), 8)

    #cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
    img2 = cv2.rectangle(img, roi_box[0], roi_box[1], (0,0,255), 3)
    x_1 = (box[0][0])
    y_1 = (box[0][1])
    x_2 = (box[2][0])
    y_2 = (box[2][1])
    
    
    # #print(x_1)
    if ((x_1 >= b_x1) and (y_1 >= b_y1)) and ((x_2 <= b_x2) and (y_2 <= b_y2)):
        print (".....Valid_document.......")
    else:
        print("....not valid.....")
    print(x_2, y_2)
    cv2.imshow("img", img2)
    #cv2.imshow("crop", roi_img)
    cv2.imshow("thresh1", thresh1)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

print(b_x2, b_y2)
print([box])   
cap.release()
cv2.destroyAllWindows()
```
